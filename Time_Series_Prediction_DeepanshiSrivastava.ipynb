```python
import pandas as pd, numpy as np
```

```python
df = pd.read_excel('DATASET.xlsx')
df
```

```python
df.fillna(0, inplace=True)
df = df.drop([910, 911, 912, 913, 914, 915, 916, 917, 918])

col = 0
for row in range(100):
  df.iloc[col:col + 10, 0] = df.iloc[col, 0]
  col += 10
```

```python
df_test = df.loc[df['year'] == 10]
df_test
```

```python
df.columns = df.columns.str.replace('Unnamed: 0', 'Group')
df_test.columns = df_test.columns.str.replace('Unnamed: 0', 'Group')
df
```

```python
#MODEL 
from sklearn import tree
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import math
```

```python
m = tree.DecisionTreeRegressor(criterion='mse',
                    min_samples_leaf=1, max_features='auto',
                    min_weight_fraction_leaf=0.0, 
                    random_state=1)
```

```python
predicted = {'Para-9': [], 'Para-10': [], 'Para-11': [], 'Para-12': [], 'Para-13': []}
names = ['Para-9', 'Para-10', 'Para-11', 'Para-12', 'Para-13']
```

```python
rmse = []
N = 0
index = 0
for j in [10, 11, 12, 13, 14]:
  ans = 0
  count = 0
  for i in range(0, 1000, 10):
    xtrain = df.iloc[i:i+9, 1:10]
    ytrain = df.iloc[i:i+9, j]
    xtest = df.iloc[i+9, 1:10].to_numpy()
    ytest = df.iloc[i+9, j]
    m.fit(xtrain, ytrain)
    ypred = m.predict(xtest.reshape(1,-1))[0]
    predicted[names[index]].append(ypred)
    ans += (ytest - ypred)**2
    count += 1
    N += 1
  rmse.append(ans)
  index += 1
```

```python
df_pred = pd.DataFrame.from_dict(predicted)
df_pred
```

```python
for i in rmse:
  print((i/count)**0.5)
```

```python
import random
generator = []
for i in range(25):
  generator.append(random.randint(0,100))
```

```python
for i in range(5):
  plt.figure(figsize=(5, 5))
  true_value = df_pred.iloc[generator, i]
  predicted_value = df_test.iloc[generator, i + 10]
  plt.scatter(true_value, predicted_value, c='black')

  p1 = max(max(predicted_value), max(true_value))
  p2 = min(min(predicted_value), min(true_value))
  plt.plot([p1, p2], [p1, p2], 'g-')
  plt.title(names[i])
  plt.xlabel('True values', fontsize=15)
  plt.ylabel('Predicted values', fontsize=15)
  plt.axis('equal')
  plt.show()
```
